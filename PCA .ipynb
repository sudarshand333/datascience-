```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

standard_scaler = StandardScaler()
std_wn = standard_scaler.fit_transform(wn)
std_wn.shape
```

```python
# To get accuracy result comparison using Minmaxscaler


minmax = MinMaxScaler()

minmax_wn = minmax.fit_transform(wn)
minmax_wn.shape
```

```python
# Import and fit the model
from sklearn.decomposition import PCA
# PCA
pca_var = PCA()
pca_var.fit(std_wn)
```

```python
# PCA
pca = PCA(n_components=3)
pca_std = pca.fit_transform(std_wn)

# Convert to data frame
pca_std_wn = pd.DataFrame(data = pca_std, columns = ['PC1', 'PC2','PC3'])

# Shape and preview
print(pca_std_wn.shape)
pca_std_wn.head()
```

```python
plt.figure(figsize=(8,6))

# Scatterplot
plt.scatter(pca_std_wn.iloc[:,0], pca_std_wn.iloc[:,1], s=40)


plt.title('PCA plot in 2D using Strongest Principle Components')
plt.xlabel('PC1')
plt.ylabel('PC2')
```

```python
# Cumulative Variance
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
pca = PCA(n_components=3)
pca_minmax = pca.fit_transform(minmax_wn)

# Convert to data frame
pca_minmax_wn = pd.DataFrame(data = pca_minmax, columns = ['PC1', 'PC2','PC3'])

# Shape and preview
print(pca_minmax_wn.shape)
pca_minmax_wn.head()
```

```python
from sklearn.cluster import AgglomerativeClustering, KMeans
```

```python
from sklearn.metrics import silhouette_score
n_clusters = [2,3,4,5,6,7,8,9,10] # number of clusters
clusters_inertia = [] # inertia of clusters
s_scores = [] # silhouette scores

for n in n_clusters:
    KM_est = KMeans(n_clusters=n, init='k-means++').fit(pca_std_wn)
    clusters_inertia.append(KM_est.inertia_)    # data for the elbow method
    silhouette_avg = silhouette_score(pca_std_wn, KM_est.labels_)
    s_scores.append(silhouette_avg) # data for the silhouette score method
```

```python
# Instantiate a scikit-learn K-Means model. we will check for two diff hyperparameters value effect.
model = KMeans(random_state=10, max_iter=500, init='k-means++')

# Instantiate the KElbowVisualizer with the number of clusters and the metric
from yellowbrick.cluster import KElbowVisualizer
visualizer = KElbowVisualizer(model, k=(2,20), metric='silhouette', timings=False)
# Fit the data and visualize
print('Elbow Plot for Standard Scaler data')
visualizer.fit(pca_std_wn)
visualizer.show()
plt.show()
```

```python
model_kmeans = KMeans(n_clusters=3, random_state=0, init='k-means++')
y_predict_kmeans = model_kmeans.fit_predict(pca_std_wn)
y_predict_kmeans.shape
```

```python
wcss=[]
for i in range (1,9):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(pca_minmax_wn)
    wcss.append(kmeans.inertia_)

# Plot K values range vs WCSS to get Elbow graph for choosing K (no. of clusters)
plt.plot(range(1,9),wcss,color = 'black')
plt.scatter(range(1,9),wcss,color='red')
plt.title('Elbow Graph for MinMaxScaler')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```
