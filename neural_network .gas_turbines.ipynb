```python
model_data = gt[['TAT', 'CO', 'AFDP', 'AT',"TEY",'CDP', 'GTEP','TIT', ]]
model_data.head()
```

```python
continuous_feature=[feature for feature in model_data.columns if model_data[feature].dtype!='O']
print('Continuous Feature Count {}'.format(len(continuous_feature)))
```

```python
gt_standard_scaled = model_data.copy()
features = gt_standard_scaled[continuous_feature]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

gt_standard_scaled[continuous_feature] = scaler.fit_transform(features.values)
gt_standard_scaled.head()
```

```python

input_features = gt_standard_scaled.drop('TEY',axis=1)
output = gt_standard_scaled[['TEY']]
```

```python
from sklearn.model_selection import train_test_split
X, X_test, Y, y_test = train_test_split(input_features, output, test_size = 0.2, random_state = 0)
```

```python
input_features.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense
model = Sequential()
model.add(Dense(20, input_dim=7,  activation='relu')) #1st layer
model.add(Dense(10,  activation='relu')) #2nd layer
model.add(Dense(10,  activation='relu')) #3nd layer
model.add(Dense(1, activation='sigmoid')) #output layer
```

```python
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['mse'])
```

```python
model.summary()
```

```python
# Fit the model
history = model.fit(X, Y, validation_split=0.2, epochs=100, batch_size=20)
```

```python
# evaluate the model on train data
scores = model.evaluate(X, Y)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
# evaluate the model on test data
scores = model.evaluate(X_test, y_test)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
# list all data in history
history.history.keys()
```

```python
model = Sequential()
model.add( Dense( units = 50 , activation = 'relu' , kernel_initializer = 'normal', input_dim = 7)) # input layer
model.add( Dense( units = 20 , activation = 'tanh' , kernel_initializer = 'normal' )) # hidden layer
model.add( Dense( units = 1  , kernel_initializer = 'normal' )) # o/p layer

model.compile(optimizer= "adam", loss="mse", metrics= ["mae", "mse"])
```

```python
# Compile the model
model.compile(loss = "mean_squared_error",optimizer = "adam",metrics = ["mae"])
```

```python
# Fit the model

history = model.fit(X, Y, validation_split=0.2, epochs=100, batch_size=20)
```

```python
# evaluate the model on train data
scores = model.evaluate(X, Y)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
# evaluate the model on test data
scores = model.evaluate(X_test, y_test)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
