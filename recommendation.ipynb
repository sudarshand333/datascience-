#**Recommendation rule**
Problem statement. Build a recommender system by using cosine simillarties score.

```python

import pandas as pd
import numpy as np

from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation

b= pd.read_csv('/content/book (1).csv', encoding= 'latin1')
b
```

```python
b.drop(columns = 'Unnamed: 0', inplace = True)
b
```

```python
print('Number of Unique Users:',len(b["User.ID"].unique()))
print('Number of Unique books :' ,len(b["Book.Title"].unique()))
```

```python
User_Book = b.pivot_table(index= "user_id",
                             columns="title",
                             values="rating").reset_index(drop=True).fillna(0)
User_Book
```

```python
#Calculating Cosine Similarity between Users

user_sim = 1-pairwise_distances(User_Book.values , metric= "cosine")
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
user_sim_df
```

```python
user_sim_df.iloc[:5,:5]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5,0:5]
```

```python
# Set the index and column name to user Ids
user_sim_df.index = list(User_Book.index)
user_sim_df.columns = list(User_Book.index)
user_sim_df
```

```python
# Most Similar Users
user_sim_df.idxmax(axis=1)
```

```python
def give_reco(id):
    tem = list(user_sim_df.sort_values([id],ascending=False).head(100).index)
    print('similar customer ids:',tem)
    movie_list=[]
    for i in tem:
        movie_list=movie_list+list(b[b['user_id']==i]['title'])
    print('Common movies within customer',movie_list)
    return set(movie_list)-set(b[b['user_id']==id]['title'])
```
